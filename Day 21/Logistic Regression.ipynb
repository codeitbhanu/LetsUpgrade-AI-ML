```python
data = pd.read_excel("Bank_Personal_Loan_Modelling.xlsx", sheet_name=1)
```

```python
data.head()
```

```python
data.shape
```

```python
import statsmodels.api as sm
```

```python
y = pd.DataFrame(data['Personal Loan'], columns=['Personal Loan'])
```

```python
x = data.drop(['Personal Loan'], axis=1)
```

```python
X1 = sm.add_constant(x)
```

```python
Logistic = sm.Logit(y, X1)
```

```python
result = Logistic.fit()
```

```python
result.summary()
```

```python
data1 = pd.read_csv("general_data.csv")
```

```python
data1.head()
```

```python
data1.info()
```

```python
data1['Attrition'] = data1['Attrition'].map({'Yes':1, 'No': 0})
```

```python
data1.head()
```

```python
data1 = data1.drop(['EmployeeCount','EmployeeID','Over18','StandardHours'],axis=1)
```

```python
data1.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
le = LabelEncoder()
```

```python
data1 = data1.dropna()
```

```python
data1.BusinessTravel = le.fit_transform(data1.BusinessTravel)
data1.Department = le.fit_transform(data1.Department)
data1.EducationField = le.fit_transform(data1.EducationField)
data1.Gender = le.fit_transform(data1.Gender)
data1.JobRole = le.fit_transform(data1.JobRole)
data1.MaritalStatus = le.fit_transform(data1.MaritalStatus)
```

```python
data1.isnull().sum()
```

```python
x = data1.drop(['Attrition'], axis=1)
y = data1.Attrition
```

```python
x1 = sm.add_constant(x)
Log_reg = sm.Logit(y, x1)
result = Log_reg.fit()
result.summary()
```
